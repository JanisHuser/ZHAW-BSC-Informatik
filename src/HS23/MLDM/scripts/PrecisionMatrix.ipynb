```python
matrix = [
	[20, 30, 10],
	[0, 100, 0],
	[0, 20, 30]
]

tp = matrix[2][2]		# TRUE POSITIVE (correctly predicted positives)
tn = matrix[0][0]		# TRUE NEGATIVE (correctly predicted negatives)
fp = matrix[2][0]		# FALSE POSITIVE (prediction=positive, actual=negative)
fn = matrix[0][2]		# FALSE NEGATIVE (prediction=negative, actual=positive)
```

```python
# Precision

print (f"tp={tp}, tn={tn}, fp={fp}, fn={fn}")
print ("accuracy", (tp+tn)/(tp+tn+fp+fn))
print ("precision positive (default)", tp/(tp+fp))
print ("precision negative", tn/(tn+fn))
print ()
print ("recall positive (default)", tp / (tp + fn))
print ("recall negative", tn / (tn + fp))
print()

precision = tp/(tp+fp)
recall =  tp / (tp + fn)
print ("f1", 2*((precision*recall)/(precision+recall)))
```
