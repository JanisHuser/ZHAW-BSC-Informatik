```python
def has_carry(result, bit_limits):
    m = (1<<bit_limits)
    return result & m == m

def is_negative(result, bit_limits):
    m = (1<<(bit_limits-1))
    return result & m == m

def is_zero(result, bit_limits):
    m = 0
    for i in range(bit_limits):
        m += (1<<i)

    return (result & m) == 0

def has_overflow(result,a,b, bit_limits):
    msb = (1<<(bit_limits-1))
    
    msb_a = (msb & a) == msb
    msb_b = (msb & b) == msb

    if not msb_a and not msb_b:
        return True
    
    return msb_a == msb_b
    
def limit(v, bit_limits):
    m = 0
    for i in range(bit_limits):
        m += (1<<i)
        
    return v & m
def adds(a,b, bit_limits=32):
    result = a+b
    
    flags = {
        "result": hex(limit(result, bit_limits)),
        "N": is_negative(result, bit_limits),
        "Z": is_zero(result, bit_limits),
        "C":  has_carry(result, bit_limits),
        "V": has_overflow(result, a, b, bit_limits)
    }
    print (flags)

    if not RET_FLAGS:   
        return limit(result, bit_limits)
    
    return (limit(result, bit_limits), flags)

def subs(a,b, bit_limits=32):
    result = a+((~b)+1)
    flags = {
        "result": hex(limit(result, bit_limits)),
        "N": is_negative(result, bit_limits),
        "Z": is_zero(result, bit_limits),
        "C": not has_carry(result, bit_limits),
        "V": not has_overflow(result, a, b, bit_limits)
    }

    print (flags)
    
    if not RET_FLAGS:   
        return limit(result, bit_limits)
    
    return (limit(result, bit_limits), flags)


def movs(v, bit_limits=32):
    result = v
    flags = {
        "result": hex(limit(result, bit_limits)),
        "N": is_negative(result, bit_limits),
        "Z": is_zero(result, bit_limits),
        "C": False,
        "V": False
    }

    print (flags)
    
    if not RET_FLAGS:   
        return limit(result, bit_limits)
    
    return (limit(result, bit_limits), flags)
    
def ldr(v, bit_limits=32):
    result = v
    
    flags = {
        "result": hex(limit(result, bit_limits)),
        "N": is_negative(result, bit_limits),
        "Z": is_zero(result, bit_limits),
        "C": False,
        "V": False
    }

    
    print (flags)
    
    if not RET_FLAGS:   
        return limit(result, bit_limits)
    
    return (limit(result, bit_limits), flags)


def lsrs(value, n, bit_limits=32):
    

    shift_values = []
    
    result = value
    for i in range(n):
        shift_values.append(result & 0x01)
        result = result >> 1
        
    carry = False
    if 1 in shift_values:
        carry = True
    
    flags = {
        "result": hex(limit(result, bit_limits)),
        "N": is_negative(result, bit_limits),
        "Z": is_zero(result, bit_limits),
        "C": carry,
        "V": False
    }

    print (flags)
    
    if not RET_FLAGS:   
        return limit(result, bit_limits)
    
    return (limit(result, bit_limits), flags)

def rors(value, n, bit_limits=32):
    shift_values = []
    
    result = value
    for i in range(n):
        shift_values.append(result & 0x01)
        result = result >> 1
        
    for i in range(len(shift_values)):
        s = shift_values[i]
        result |= (s << (31 - i))
    
    flags = {
        "result": hex(limit(result, bit_limits)),
        "N": is_negative(result, bit_limits),
        "Z": is_zero(result, bit_limits),
        "C": 1 in shift_values,
        "V": False
    }

    print (flags)
    
    if not RET_FLAGS:   
        return limit(result, bit_limits)
    
    return (limit(result, bit_limits), flags)

def mov(v, bit_limits):
    print ("NO FLAGS UPDDATED")
    return limit(v, bit_limits)

def muls(a, b, bit_limits):
    a = limit(a, bit_limits)
    b = limit(b, bit_limits)

    result = a * b
    limited = limit(result, bit_limits)

    flags = {
        "result": hex(limited),
        "N": is_negative(result, bit_limits),
        "Z": is_zero(result, bit_limits),
        "C": False,
        "V": result > limited
    }
    
    if not RET_FLAGS:   
        return limited
    
    return (limited, flags)


def cmp(Rn, Rm, bit_limits=32):
    _, flags = subs(Rn, Rm, bit_limits)

    return flags

def cmn(Rn, Rm, bit_limits=32):
    _, flags = adds(Rn, Rm, bit_limits)
    return flags


def tst(Rn, Operand2):
    result = Rn & Operand2

    flags = {
        'Z': result == 0,  # Zero flag is True if result is zero,
        'C': False,
        'N': False,
        'V': False
    }

    return flags
```
