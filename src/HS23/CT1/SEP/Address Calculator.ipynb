```python
address = 0
def add_area(name, size, lowest_address=0):
    global address
    if lowest_address > address:
        address = lowest_address
        
        
    to_address = address + size -1
    print(f"{name} -> from: {hex(address)}, to: {hex(to_address)}")
    
    address = to_address + 1
    
    
add_area("CODE", 1024, 0x20000700)
add_area("STACK", 256)
add_area("DATA", 512)
```

```python

def dcb(count=1):
	# BYTE
	return 1*count

def dcw(count=1):
	# WORD
	return 2*count

def dcd(count=1):
	# DOUBLE WORD
	return 4*count

def space(size):
	return size
```

```python
# USAGE
# var	DCD 0x0 			=> dcd(1)
# var	DCD 0x0, 0x0 		=> dcd(2)
# var	DCW 0x0 			=> dcw(1)
# var	DCW 0x0, 0x0 		=> dcw(2)
# var	DCB 0x0 			=> dcb(1)
# var	DCB 0x0, 0x0 		=> dcb(2)
# var	SPACE	16			=> space(16)
size = 0
size += dcd(1)
size += dcd(1)
size += dcd(2)


print ("allocation size [bytes]", size)
```
