# KNN - K Nearest Neighbour

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor

# Generate some sample data
X = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

# Creating KNN Regression Model
knn_model = KNeighborsRegressor(n_neighbors=3)  # You can adjust the value of n_neighbors
knn_model.fit(X, y)

# Make predictions
X_pred = np.arange(0, 60, 1).reshape((-1, 1))
y_pred = knn_model.predict(X_pred)

# Plotting the data points
plt.scatter(X, y, color='black', label='Training Data')

# Plotting the KNN Regression line
plt.plot(X_pred, y_pred, color='blue', label='KNN Regression')

plt.title('K-Nearest Neighbors Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

# Display the plot
plt.show()
```

## Advantages
**Advantages of K-Nearest Neighbors (KNN):**

1. **Simple and Intuitive:**
   - KNN is easy to understand and implement. It's a straightforward algorithm that doesn't require complex assumptions about the underlying data distribution.

2. **No Training Phase:**
   - KNN is a lazy learner, meaning it doesn't have a training phase. The model is built during the prediction phase by memorizing the entire training dataset. This can be advantageous for dynamic datasets.

3. **Versatile:**
   - KNN can be applied to both regression and classification problems. It can handle a variety of data types, including numeric and categorical features.

4. **Adapts to Local Patterns:**
   - KNN adapts well to local patterns in the data. It can capture complex relationships, including nonlinear patterns, without assuming a specific form for the underlying model.

5. **Robust to Outliers:**
   - KNN can be robust to outliers, as it considers the majority class among the k-nearest neighbors. Outliers may have less impact on predictions if they are not part of the majority in the local neighborhood.

6. **No Training Time:**
   - Since KNN doesn't involve a training phase, the model can quickly adapt to changes in the data. This makes it suitable for scenarios where the relationship between features and the target variable is expected to change over time.


## Disadvantages

1. **Computational Complexity:**
   - The prediction phase in KNN can be computationally expensive, especially for large datasets or high-dimensional feature spaces. Finding the k-nearest neighbors requires distance calculations for each data point.

2. **Memory Usage:**
   - KNN requires storing the entire training dataset in memory. For large datasets, this can lead to high memory usage.

3. **Sensitivity to Feature Scale:**
   - KNN is sensitive to the scale of features. Features with larger scales can dominate the distance calculations. Normalizing or standardizing features is often necessary.

4. **Curse of Dimensionality:**
   - In high-dimensional spaces, the distance between data points tends to increase, and the concept of "closeness" becomes less meaningful. This is known as the curse of dimensionality and can affect the performance of KNN.

5. **Choice of Distance Metric:**
   - The choice of distance metric (e.g., Euclidean, Manhattan) can impact the performance of KNN. The optimal metric depends on the characteristics of the data.

6. **Parameter Sensitivity:**
   - The performance of KNN is sensitive to the choice of the `k` parameter. Selecting an appropriate value for `k` requires careful consideration and experimentation.

7. **Imbalanced Datasets:**
   - In datasets with imbalanced class distributions, KNN may be biased towards the majority class, especially when using a small value of `k`.

8. **Not Suitable for Large Datasets:**
   - Due to its computational complexity, KNN may not be suitable for large datasets, and the prediction time can be significantly high.
