```python
import matplotlib.pyplot as plt

# Daten: Datum und Anzahl der Arbeitseinheiten in jeder Phase
dates = ["01.01.2024", "02.01.2024", "03.01.2024", "04.01.2024", "05.01.2024"]
to_do = [10, 12, 15, 13, 10]
in_progress = [5, 6, 4, 7, 8]
review = [2, 1, 3, 2, 3]
done = [3, 5, 7, 8, 9]

# Kumulative Summen berechnen
cumulative_to_do = [sum(to_do[:i+1]) for i in range(len(to_do))]
cumulative_in_progress = [cumulative_to_do[i] + sum(in_progress[:i+1]) for i in range(len(in_progress))]
cumulative_review = [cumulative_in_progress[i] + sum(review[:i+1]) for i in range(len(review))]
cumulative_done = [cumulative_review[i] + sum(done[:i+1]) for i in range(len(done))]

# CFD Diagramm erstellen
plt.fill_between(dates, cumulative_to_do, color='skyblue', alpha=0.5, label='To Do')
plt.fill_between(dates, cumulative_to_do, cumulative_in_progress, color='sandybrown', alpha=0.5, label='In Progress')
plt.fill_between(dates, cumulative_in_progress, cumulative_review, color='lightgreen', alpha=0.5, label='Review')
plt.fill_between(dates, cumulative_review, cumulative_done, color='lightcoral', alpha=0.5, label='Done')

plt.xlabel('Datum')
plt.ylabel('Kumulative Anzahl der Arbeitseinheiten')
plt.title('Cumulative Flow Diagram (CFD)')
plt.legend()
plt.show()
```
